```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('house.csv')
```

```python
df
```

```python
df.info()
```

```python
df = df[['GrLivArea', 'YearBuilt', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'OverallQual', 'Neighborhood', 'SalePrice']]
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['Neighborhood'], drop_first=True).astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = df.drop(columns=('SalePrice'))
scaled_data = scaler.fit_transform(features)
```

```python
x = scaled_data
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
lin_model = LinearRegression()
```

```python
lin_model.fit(x_train, y_train)
```

```python
y_pred = lin_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
import joblib

joblib.dump(lin_model, 'model.pkl')
joblib.dump(scaler, 'scaler.pkl')
```
